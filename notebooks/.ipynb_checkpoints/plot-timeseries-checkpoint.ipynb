## 1. surface flux timeseries - $\langle w'\theta'\rangle_0$, $\langle w'q'\rangle_0$, $ u^\star_0$

```python
fig, axes = plt.subplots(3, 1, figsize=(7, 7), sharex=True)

ds_run1.wtheta.plot(ax = axes[0], label = 'run 1')
ds_run1.wq.plot(ax = axes[1])
ds_run1.ustar.plot(ax = axes[2])

ds_run2.wtheta.plot(ax = axes[0], label = 'run 2')
ds_run2.wq.plot(ax = axes[1])
ds_run2.ustar.plot(ax = axes[2])

axes[0].legend()
axes[0].grid(linestyle = '--')
axes[1].grid(linestyle = '--')
axes[2].grid(linestyle = '--')
plt.savefig('./notebooks/variable-sst-DALES-analysis/figures/surface-flux-timeseries.png')
```
