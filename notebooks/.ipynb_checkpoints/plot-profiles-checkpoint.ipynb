# DALES plot profiles

```python
import xarray as xr
import matplotlib.pyplot as plt
import cmocean as cmo
import numpy as np
```

```python
# Load data
output_directory = '/scratch/fy29/mxr581/dales/run1'
filename = 'profiles.001_055.run_1.nc'
file_path = f"{output_directory}/{filename}"
ds_run1 = xr.open_dataset(file_path)

output_directory = '/scratch/jk72/mxr581/dales/run2'
filename = 'profiles.001_054.run_2.nc'
file_path = f"{output_directory}/{filename}"
ds_run2 = xr.open_dataset(file_path)

output_directory = '/scratch/fy29/mxr581/dales/run3'
filename = 'profile.001_054.run_3.nc'
file_path = f"{output_directory}/{filename}"
ds_run3 = xr.open_dataset(file_path)
```

```python
#ds_run1.time[0]
```

## 1. plot profiles of velocity components $u$ and $v$

```python
fig, axes = plt.subplots(1, figsize=(3, 3))

# Restrict zt to < xxx m
z_max = 100
t_start = np.datetime64('2020-02-01T10:00:00')
t_end = t_start + np.timedelta64(10, 'h')

u_timeavg_run1 = ds_run1.u.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))
v_timeavg_run1 = ds_run1.v.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))
spd_run1 = (u_timeavg_run1**2+v_timeavg_run1**2)**0.5

u_timeavg_run2 = ds_run2.u.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))
v_timeavg_run2 = ds_run2.v.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))
spd_run2 = (u_timeavg_run2**2+v_timeavg_run2**2)**0.5

u_timeavg_run3 = ds_run3.u.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))
v_timeavg_run3 = ds_run3.v.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))
spd_run3 = (u_timeavg_run3**2+v_timeavg_run3**2)**0.5

spd_run1.plot(ax=axes, y = 'zt', color = 'salmon', 
              #yscale = 'log', 
              marker = '.', label = r'run 1 $L/\lambda = 1$')
spd_run2.plot(ax=axes, y = 'zt', color = 'teal', 
              #yscale = 'log', 
              linestyle = '-', marker = '.',  label = r'run 1 $L/\lambda = 2$')

spd_run3.plot(ax=axes, y = 'zt', color = 'midnightblue',  
              #yscale = 'log', 
              linestyle = '-', marker = '.',  label = r'run 1 $L/\lambda = 3$')
axes.legend()

axes.set(xlim = [0, 4.5], ylim = [0, z_max], xlabel = 'speed (m/s)', ylabel = 'height (m)')
plt.savefig('./notebooks/DALES-analysis/figures/velocity-profiles.png')
```

## 2. plot profiles of virtual potential temp $\theta_v$

```python
fig, axes = plt.subplots(1, figsize=(3, 3))

z_max = 100

t_start = np.datetime64('2020-02-01T00:00:00')
t_end = t_start + np.timedelta64(1, 'h')

thv_timeavg_run1 = ds_run1.thv.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))
thv_timeavg_run2 = ds_run2.thv.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))
thv_timeavg_run3 = ds_run3.thv.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))

thv_timeavg_run1.plot(ax=axes, y = 'zt', color = 'salmon', 
            label = 'hour 1')

thv_timeavg_run2.plot(ax=axes, y = 'zt', color = 'teal', 
              linestyle = '-')

thv_timeavg_run2.plot(ax=axes, y = 'zt', color = 'midnightblue',  
              linestyle = '-')

t_start = np.datetime64('2020-02-02T00:00:00')
t_end = t_start + np.timedelta64(1, 'h')

thv_timeavg_run1 = ds_run1.thv.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))
thv_timeavg_run2 = ds_run2.thv.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))
thv_timeavg_run3 = ds_run3.thv.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))

thv_timeavg_run1.plot(ax=axes, y = 'zt', color = 'salmon', 
              linestyle = '--', label = 'hour 25')

thv_timeavg_run2.plot(ax=axes, y = 'zt', color = 'teal', 
              linestyle = '--')

thv_timeavg_run2.plot(ax=axes, y = 'zt', color = 'midnightblue',  
              linestyle = '--')

t_start = np.datetime64('2020-02-03T00:00:00')
t_end = t_start + np.timedelta64(1, 'h')

thv_timeavg_run1 = ds_run1.thv.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))
thv_timeavg_run2 = ds_run2.thv.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))
thv_timeavg_run3 = ds_run3.thv.sel(time=slice(t_start, t_end)).mean('time').sel(zt=slice(0, z_max))

thv_timeavg_run1.plot(ax=axes, y = 'zt', color = 'salmon', 
              linestyle = '-.', label = 'hour 49')

thv_timeavg_run2.plot(ax=axes, y = 'zt', color = 'teal', 
              linestyle = '-.')

thv_timeavg_run2.plot(ax=axes, y = 'zt', color = 'midnightblue',  
              linestyle = '-.')

axes.legend()
axes.set(ylim = [0, z_max], xlabel = r'$\theta_v$ (K)', ylabel = 'height (m)')

plt.savefig('./notebooks/DALES-analysis/figures/temp-profiles.png')
```

## 3. Plot virtual potential temp $\theta_v$ hovmoller

```python
# Set up a wide figure: 3 subplots side by side
fig, axes = plt.subplots(3, 1, figsize=(7, 6), sharex=True)

# Restrict zt to < 1500 m
zt_max = 1500

# Compute common color limits across runs
vmin = ds_run1.thv.sel(zt=slice(0, zt_max)).min()
vmax = ds_run1.thv.sel(zt=slice(0, zt_max)).max()

# --- Run 1 ---
ds_run1.thv.sel(zt=slice(0, zt_max)).plot(
    x='time', y='zt', ax=axes[0], cmap='cmo.thermal',
     vmin=vmin, vmax=vmax,
    cbar_kwargs={'label': 'θv (K)'}
)
axes[0].set(title = 'Run 1', xlabel = '', ylabel = 'Height (m)')

# --- Run 2 ---
ds_run2.thv.sel(zt=slice(0, zt_max)).plot(
    x='time', y='zt', ax=axes[1], cmap='cmo.thermal',
     vmin=vmin, vmax=vmax,
    cbar_kwargs={'label': 'θv (K)'}
)
axes[1].set(title = 'Run 2', xlabel = '', ylabel = 'Height (m)')

#--- Run 3 (optional) ---
ds_run3.thv.sel(zt=slice(0, zt_max)).plot(
    x='time', y='zt', ax=axes[2], cmap='cmo.thermal',
     vmin=vmin, vmax=vmax,
    cbar_kwargs={'label': 'θv (K)'}
)
axes[2].set(title = 'Run 3', xlabel = 'time', ylabel = 'Height (m)')

plt.tight_layout()
plt.show()
```

```python
# Select θv at ~100 m height for each run
thv1 = ds_run1.thv.sel(zt=100, method='nearest')
thv2 = ds_run2.thv.sel(zt=100, method='nearest')
thv3 = ds_run3.thv.sel(zt=100, method='nearest')

# Create figure
plt.figure(figsize=(8, 4))

# Plot time series
plt.plot(thv1['time'], thv1, label='Run 1')
plt.plot(thv2['time'], thv2, label='Run 2')
plt.plot(thv3['time'], thv3, label='Run 3')

# Formatting
plt.xlabel('Time')
plt.ylabel('θv (K)')
plt.title('Virtual Potential Temperature at 100 m')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
